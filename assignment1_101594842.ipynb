```python
"""
Author: Wasifa Hossain
Assignment:#1
"""

gym_member = "Alex Alliton" # string
preferred_weight_kg = 20.5 # float
highest_reps = 25 # integer
membership_active = True # boolean



# workout_stats dictionary where keys are friend names (strings), and values are tuples (immutable sequence of integers)
workout_stats = {
    "Alex": (30, 45, 20),     # yoga, running, weightlifting
    "Jamie": (30, 35, 30),
    "Taylor": (25, 45, 40)
}

# total workout minutes for each friend
for friend, minutes in list(workout_stats.items()):
    yoga, running, weights = minutes
    total_minutes = sum(minutes)
    workout_stats[f"{friend}_Total"] = total_minutes

# 2D nested list (strings and integers) containing workout minutes for each friend
workout_list = [
    workout_stats["Alex"],
    workout_stats["Jamie"],
    workout_stats["Taylor"]
]

# yoga and running minutes for all friends
print("Yoga and Running minutes for all friends:")
for workout in workout_list:
    print(workout[:2])

# weightlifting minutes for the last two friends
print("\nWeightlifting minutes for the last two friends:")
for workout in workout_list[-2:]:
    print(workout[2])


# Check if total workout minutes >= 120
for friend in ["Alex", "Jamie", "Taylor"]:
    if workout_stats[f"{friend}_Total"] >= 120:
        print(f"Great job staying active, {friend}!")


# User input feature
friend_name = input("\nEnter a friend's name: ")

if friend_name in workout_stats and isinstance(workout_stats[friend_name], tuple):
    yoga, running, weightlifting = workout_stats[friend_name]
    total = workout_stats[f"{friend_name}_Total"]
    print(f"\nWorkout details for {friend_name}:")
    print(f"Yoga: {yoga} minutes")
    print(f"Running: {running} minutes")
    print(f"Weightlifting: {weightlifting} minutes")
    print(f"Total: {total} minutes")
else:
    print(f"Friend {friend_name} not found in the records.")


# the highest and the lowest total workout minutes
totals = {
    friend: workout_stats[f"{friend}_Total"]
    for friend in ["Alex", "Jamie", "Taylor"]
}

highest_friend = max(totals, key=totals.get)
lowest_friend = min(totals, key=totals.get)

print(f"\nFriend with the highest total workout minutes: {highest_friend}")
print(f"Friend with the lowest total workout minutes: {lowest_friend}")
```
